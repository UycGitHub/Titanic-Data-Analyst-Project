```python
def variables(variable):
  cat = train_data[variable]
  count = cat.value_counts()
  plt.figure(figsize=(9,3))
  plt.bar(count.index,count)
  plt.xticks(count.index, count.index.values)
  plt.ylabel("frekans")
  plt.title(variable)
  plt.show()
  print(variable,count)
```

```python
category1 = ["Survived", "Sex", "Pclass", "Embarked", "SibSp", "Parch"]
```

```python
for i in category1:
  variables(i)
```

```python
category2 = ["Cabin", "Name", "Ticket",]
for i in category2:
  print(train_data[i].value_counts)
```

```python
def hist(variable):
  plt.figure(figsize=(9,3))
  plt.hist(train_data[variable])
  plt.xlabel(variable)
  plt.ylabel("Frekans")
  plt.title(variable)
  plt.show()
```

```python
numeric = ["Fare", "Age", "PassengerId"]

for i in numeric:
  hist(i)
```
